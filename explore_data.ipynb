```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Customers dataset

<strong> customer_id </strong> : Khóa liên kết với tập dữ liệu customer. Mỗi đơn hàng có một customer_id duy nhất.

<strong> customer_unique_id </strong>: Khóa định danh duy nhất của một khách hàng

<strong> customer_zip_code_prefix </strong>: 5 kí tự đầu tiên của mã zip của khách hàng

<strong> customer_city </strong>: Thành phố nơi khách hàng sinh sống

<strong> customer_state </strong>:  Bang nơi khách hàng sinh sống

```python
customers_df = pd.read_csv('./data/olist_customers_dataset.csv')
customers_df.head()
```

```python
print('+ Shape of dataframe:', customers_df.shape)
print('+ Data type of each column:')
customers_df.dtypes
```

```python
unique_customers = customers_df.drop_duplicates(subset='customer_unique_id', ignore_index=True)

plt.figure(figsize=(14,8))
ax = unique_customers['customer_state'].value_counts().head(10).plot(kind='bar', edgecolor='black')                                                   
plt.title('Top 10 States with the Highest Number of Unique Customers', fontsize=16, fontweight='bold')
plt.xlabel('State', fontsize=14)
plt.ylabel('Number of Customers', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=10)

plt.show()
```

### Orders dataset

<strong> order_id </strong>: Khóa định danh duy nhất của từng đơn hàng.

<strong> customer_id </strong>: Khóa liên kết với tập dữ liệu customer. Mỗi đơn hàng có một customer_id duy nhất.

<strong> order_status </strong>: Tham chiếu đến trạng thái đơn hàng (đã giao, đang vận chuyển, ...).

<strong> order_purchase_timestamp </strong>: Hiển thị dấu thời gian mua hàng.

<strong> order_approved_at </strong>: Hiển thị dấu thời gian phê duyệt thanh toán.

<strong> order_delivered_carried_date </strong>: Hiển thị thời gian khi đơn hàng được bàn giao cho đối tác vận chuyển.

<strong> order_delivered_customer_date  </strong>: Hiển thị thời gian thực tế giao đơn hàng cho khách hàng.

<strong> order_estimated_delivery_date </strong>: Hiện thị thời gian giao hàng ước tính mà khách hàng được thông báo tại thời điểm mua hàng.

```python
orders_df = pd.read_csv('./data/olist_orders_dataset.csv')
orders_df.head()
```

```python
print('+ Shape of dataframe:', orders_df.shape)
print('+ Data type of each column:')
orders_df.dtypes
```

```python
plt.figure(figsize=(14,8))
ax = orders_df['order_status'].value_counts().plot(kind='bar', edgecolor='black')                                                   
plt.title('Number of Orders By Order Status', fontsize=16, fontweight='bold')
plt.xlabel('Order Status', fontsize=14)
plt.ylabel('Numbers of Orders', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
delayed_orders = orders_df.loc[orders_df['order_delivered_customer_date'] > orders_df['order_estimated_delivery_date'], :]
on_time_orders = orders_df.loc[~orders_df.index.isin(delayed_orders.index), :]

status_counts = [len(delayed_orders), len(on_time_orders)]
status_labels = ['Delayed', 'On Time']

# Tạo legend với số lượng thực tế
legend_labels = [f'{label}: {count}' for label, count in zip(status_labels, status_counts)]

plt.figure(figsize=(14, 8))
plt.pie(status_counts, labels=status_labels, autopct='%1.1f%%', startangle=90, colors=['red', 'green'], explode=(0, 0.1), shadow=True)
plt.title('Percent of Delayed vs On Time Orders', fontsize=16, fontweight='bold')
plt.legend(legend_labels, loc='upper right')
plt.show()
```

<strong> product_category_name_english </strong>: Danh mục của sản phẩm (tiếng Anh)

```python
merge_df = pd.merge(
        products_df, 
        product_category_name, 
        on='product_category_name', 
        how='inner'
    )
    dim_products = merge_df[[
            'product_id', 
            'product_category_name_english']]
```

```python
    merge_df = pd.merge(
            orders_df, 
            order_payments_df, 
            on='order_id', 
            how='inner'
        )
    merge_df = pd.merge(
            merge_df, 
            order_items_df, 
            on='order_id', 
            how='inner'
        )
    fact_sales = merge_df[[
            'order_id', 
            'customer_id', 
            'seller_id', 
            'order_purchase_timestamp', 
            'product_id', 
            'price',
            'payment_value',
            'payment_type', 
            'order_status'
        ]]
```

```python
    successful_orders = fact_sales
    successful_orders['order_purchase_timestamp'] = pd.to_datetime(successful_orders['order_purchase_timestamp']).dt.date
    
    daily_sales_products = successful_orders.groupby([
        'order_purchase_timestamp', 
        'product_id']).agg({
                        'payment_value': 'sum',
                        'order_id': pd.Series.nunique  
    }).reset_index()

    daily_sales_products = daily_sales_products.rename(columns={
        'order_purchase_timestamp': 'daily',
        'payment_value': 'sales',
        'order_id': 'bills'
    })
    
    daily_sales_products['daily'] = pd.to_datetime(daily_sales_products['daily'])
    daily_sales_products['monthly'] = daily_sales_products['daily'].dt.strftime("%Y-%m")

    merge_df = pd.merge(
        daily_sales_products,
        dim_products[['product_id', 'product_category_name_english']],
        on='product_id',
        how='inner'
    )

    grouped_df = merge_df.groupby(['monthly', 'product_category_name_english']).agg({
        'sales': 'sum',
        'bills': 'sum'
    }).reset_index()
    
    grouped_df = grouped_df.rename(columns = {
        'sales': 'total_sales',
        'bills': 'total_bills',
        'product_category_name_english': 'category'
    })
    grouped_df['values_per_bills'] = grouped_df['total_sales'] / grouped_df['total_bills']
    monthly_sales_categories = grouped_df[['monthly', 'category', 'total_sales', 'total_bills', 'values_per_bills']]
   
```
